```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('salaries_dataset.xlsx')
```

```python
df.head()
```

```python
def convert_str_to_list(string: str | float) -> list | float:
    # avoid null values
    if not isinstance(string, float):
        return string.replace('[','')\
        .replace(']','')\
        .replace("'",'')\
        .split(', ')
    else:
        return string
```

```python
df.loc[:, 'job_skills'] = df.job_skills.apply(convert_str_to_list)
```

```python
df = df.explode('job_skills')
```

```python
df.job_schedule_type.unique()
```

```python
def clean_job_schedule(string: str | float) -> list | float:
    if not isinstance(string, float):
        clean_string = string.replace(', and ', ', ')\
        .replace(' and ', ', ')\
        .strip()

        return clean_string.split(', ')
    else:
        return string
```

```python
df.loc[:, 'job_schedule_type'] = df.job_schedule_type.apply(clean_job_schedule)
```

```python
df = df.explode('job_schedule_type')
```

```python
df.dtypes
```

```python
df = df.sort_values('job_posted_date').reset_index(drop = True)
```

```python
df.to_csv('salaries_dataset.csv', index=False)
```
