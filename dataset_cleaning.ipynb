```python
def convert_str_to_list(string: str | float) -> list | float:
    # avoid null values
    if not isinstance(string, float):
        return string.replace('[','')\
        .replace(']','')\
        .replace("'",'')\
        .split(', ')
    else:
        return string
```

```python
df.loc[:, 'job_skills'] = df.job_skills.apply(convert_str_to_list)
```

```python
df = df.explode('job_skills')
```

```python
def clean_job_schedule(string: str | float) -> list | float:
    if not isinstance(string, float):
        clean_string = string.replace(', and ', ', ')\
        .replace(' and ', ', ')\
        .strip()

        return clean_string.split(', ')
    else:
        return string
```

```python
df.loc[:, 'job_schedule_type'] = df.job_schedule_type.apply(clean_job_schedule)
```

```python
df = df.explode('job_schedule_type')
```

```python
df = df.sort_values('job_posted_date').reset_index(drop = True)
```

```python
# Crear función para convertir nombres a códigos ISO
def get_country_iso_code(country_name):
    try:
        return pycountry.countries.search_fuzzy(country_name)[0].alpha_3
    except:
        return None
```

```python
countries_dict = {k:get_country_iso_code(k) for k in df.job_country.unique()}
```

```python
countries_dict
```

```python
df['country_code'] = df.job_country.apply(lambda x: countries_dict[x])
```

```python
list(df.columns).index('job_country')
```

```python
new_ordered_cols = list(df.columns[:11]) + ['country_code'] + list(df.columns[11:-1])
```

```python
df = df[new_ordered_cols]
```

```python
df.head()
```

```python
job_positions_by_country = (df.groupby(['job_country', 'country_code'])
                            .size()
                            .to_frame()
                            .rename(columns = {0: 'count'})
                            .reset_index()
)
job_positions_by_country.head()
```

```python
fig = px.choropleth(
    job_positions_by_country,
    locations="country_code",
    color="count",
    hover_name="job_country",
    title="Ventas por País"
)
fig.show()
```
